```python
'''     
                                                2. Matrix Modification

Write a program that reads a matrix from the console and changes its values. On the first line, you will get the matrix's
rows - N. You will get elements for each column on the following N lines, separated with a single space. You will be
receiving commands in the following format:

        . "Add {row} {col} {value}" - Increase the number at the given coordinates with the value.
        . "Subtract {row} {col} {value}" - Decrease the number at the given coordinates by the value.

If the coordinate is invalid, you should print "Invalid coordinates". A coordinate is valid if both of the given
indexes are in range [0; len() - 1].

When you receive "END", you should print the matrix and stop the program.

Examples


Input                   Output
3                       6 2 3
1 2 3                   4 3 6
4 5 6                   7 8 9 
7 8 9                   
Add 0 0 5
Subtract 1 1 2 
END


Input                   Output
4                       Invalid coordinates
1 2 3 4                 Invalid coordinates
5 6 7 8                 -41 2 3 4
8 7 6 5                  5 6 7 8
4 3 2 1                  8 7 6 5
Add 4 4 100              4 3 2 101
Add 3 3 100    
Subtract -1 -1 42
Subtract 0 0 42
END




'''

n = int(input('Enter n of rows: '))
matrix = []
command = ''
lst = []

print('User input matrix below: ')
for _ in range(n):
    row = input('Enter row values: ').split(' ')
    matrix.append(row)
    print(row)


while command != 'END':
    command = input('Enter command: ')

    if command != 'END':
        lst += [command]
        print('User Command: ', command)
    else:
        break

for i in range(len(lst)):
    
    target = lst[i]
    
    if 'Subtract' in target:
        row,col = int(str(target).split(' ')[1]), int(str(target).split(' ')[2])

        if row <= len(matrix)-1 and col <= len(matrix[0])-1:
            number = int(str(target).split(' ')[3])
            old_number = int(matrix[row][col])
            matrix[row][col] = str(old_number  - number)



    elif 'Add' in target:
        row,col = int(str(target).split(' ')[1]), int(str(target).split(' ')[2])

        if row <= len(matrix)-1 and col <= len(matrix[0])-1:
            number = int(str(target).split(' ')[3])
            old_number = int(matrix[row][col])
            matrix[row][col] = str(number + old_number)

print('\nResult Matrix: ')
for x in matrix:
    print(x)


        

```
