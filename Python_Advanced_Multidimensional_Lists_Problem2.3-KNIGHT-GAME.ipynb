```python
'''''   

                                                                3. Knight Game



Chess is the oldest game, but it is still popular these days. It will be used only one
chess piece for this task - the Knight.

A chess knight has 8 possible moves it can make, as illustrated. It can move to
the nearest square but not on the same row, column, or diagonal. (e.g., it can
move two squares horizontally, then one square vertically, or it can move one
square horizontally then two squares vertically - i.e., in an "L" pattern.)

The knight game is played on a board with dimensions N x N.
You will receive a board with "K" for knights and "0" for empty cells. Your task
is to remove knights until no knights that can attack one another with one
move are left.

Input

    . On the first line, you will receive integer N - the size of the board
    . On the following N lines, you will receive strings with "K" and "0"

Output
    . Print a single integer with the minimum number of knights that need to be removed


Constraints
    . The size of the board will be 0 < N < 30
    . Time limit: 0.3 sec. Memory limit: 16 MB

Examples:

Input       Output
5           1
0 K 0 K 0
K 0 0 0 K
0 0 K 0 0
K 0 0 0 K
0 K 0 K 0


Input       Output
2           0
K  K
K  K



Input                   Output
8                       12
0 K 0 K K K 0 0
0 K 0 0 K K K K
0 0 K 0 0 0 0 K
K K K K K K 0 K
K 0 K 0 0 0 0 K
K K 0 0 0 0 0 K
0 0 K 0 K 0 0 0
0 0 0 K 0 0 K K
'''

n = int(input('Enter matrix size dimension: '))

matrix = [input().split(' ') for _ in range(n)]
print('User input matrix: ')
for x in matrix:
    print(x)

def k_in_range(row,col, matrix):
    if matrix[row][col] == 'K':
        return True


def k_attack_counter(K, K_position, matrix):
    k_row, k_col= int(K_position.split(',')[0]), int(K_position.split(',')[1])
    matrix_row, matrix_col = len(matrix)-1, len(matrix[0])-1
    k_count = 0
 
    if 0 <= k_row - 2 and 0 <= k_col - 1 and k_in_range(k_row-2, k_col-1, matrix) == True and K == 'K':
        k_count += 1

    if 0 <= k_row - 2 and k_col  + 1 <=  matrix_col and k_in_range(k_row-2, k_col+1, matrix) == True and K == 'K':
        k_count += 1

    if 0 <= k_row - 1 and 0 <= k_col - 2 and k_in_range(k_row-1, k_col-2, matrix) == True and K == 'K':
        k_count += 1

    if 0 <= k_row - 1 and k_col + 2 <=  matrix_col and k_in_range(k_row-1, k_col+2, matrix) == True and K == 'K':
        k_count += 1

    if  k_row + 2 <= matrix_row and 0 <= k_col - 1 and k_in_range(k_row+2, k_col-1, matrix) == True and K == 'K':
        k_count += 1

    if k_row + 2 <= matrix_row and k_col  + 1 <=  matrix_col and k_in_range(k_row+2, k_col+1, matrix) == True and K == 'K':
        k_count += 1

    if k_row + 1 <= matrix_row and 0 <= k_col - 2 and k_in_range(k_row+1, k_col-2, matrix) == True and K == 'K':
        k_count += 1

    if k_row + 1 <= matrix_row and k_col + 2 <=  matrix_col and k_in_range(k_row+1, k_col+2, matrix) == True and K == 'K':
        k_count += 1

    else:
        pass

    
    return k_count
    


lst = []
 
position = ''
result_matrix =  []
 
result = ''
flag = False
k_removed = 0

while flag != True:
    scores = []
    coordinates = []
    counter = -1
    for row in matrix:
        counter += 1
        for index in range(len(row)):
            position = f'{counter},{index}'
            element = matrix[counter][index]
            result = k_attack_counter(element, position, matrix)
            if result > 0:
                scores += [f'{result}']
                coordinates += [f'{counter},{index}']

    
    
    score = 0
    coordinate = ''
    counter = -1
     



    for i in range(len(scores)):
        x = scores[i]
        y = coordinates[i]

        if int(x) > score:
            score = int(x)
            coordinate = y
     
    
    if score >= 1:
        xx,yy = int(coordinate.split(',')[0]), int(coordinate.split(',')[1])
        matrix[xx][yy] = 'x'
        k_removed += 1
        score = 0
         
    

    elif score == 0:
        flag = True
                 
print("\n\nFinal Matrix: \n'x' symbol represents removed 'X' horse\n")
for x in matrix:
    print(x)
 
print(f"\nRESULT: Total # of 'K' horses removed: {k_removed}")
        
 
       
        

           
 
    


```
